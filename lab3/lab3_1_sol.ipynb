```python
import matplotlib.pyplot as plt

def plot_sensor_data(dataframe):
    plt.figure(figsize=(12, 6))
    plt.plot(dataframe['timestamp'], dataframe['sensor_00'], label='Sensor 1')
    plt.plot(dataframe['timestamp'], dataframe['sensor_01'], label='Sensor 2')
    plt.plot(dataframe['timestamp'], dataframe['sensor_02'], label='Sensor 3')
    plt.plot(dataframe['timestamp'], dataframe['sensor_03'], label='Sensor 4')
    plt.xlabel('Tiempo')
    plt.ylabel('Valor')
    plt.title('Gráfico de Sensores 00, 01, 02, 03')
    plt.legend()
    plt.show()
```

```python
plot_sensor_data(cleaned_data)
```

#### Interpolacion

```python
data_no_nan.columns
```

```python
import numpy as np
```

```python
# Vamos a interpolar los valores 0 en las columnas de sensores porque representan valores faltantes/nulos

# Obtenemos la lista de columnas de sensores (excluyendo 'timestamp' y 'machine_status')
sensor_cols = [col for col in data_no_nan.columns if col not in ['timestamp', 'machine_status']]

# Reemplazamos los valores 0 por np.nan solo en las columnas de sensores
data_no_nan[sensor_cols] = data_no_nan[sensor_cols].replace(0, np.nan)

# Interpolamos los valores faltantes (nulos) solo en las columnas de sensores, de manera lineal
data_no_nan[sensor_cols] = data_no_nan[sensor_cols].interpolate(method='linear', limit_direction='both')

# Si quedan nulos al inicio o final (por ej. si hay nulos prolongados) puedes volver a rellenar si lo conviene:
# data_no_nan[sensor_cols] = data_no_nan[sensor_cols].fillna(method='bfill').fillna(method='ffill')
```

```python
plot_sensor_data(data_no_nan)
```
