## Frecuency Domain Analysis

```python
data_no_nan.shape
```

```python
def data_preparation_fft(df, granularity):
    # N es el número de puntos/muestras en el dataframe (número de filas)
    N = df.shape[0]

    # Se realiza la FFT (Fast Fourier Transform) a cada columna (sensor) del DataFrame.
    # axis=0 indica que la transformación se hace a lo largo de las filas (para cada sensor)
    data_fft = np.fft.fft(df.values, axis=0)
    # Se obtiene el valor absoluto de la FFT (magnitud)
    data_fft = np.abs(data_fft)
    
    # Normalización de la magnitud de la FFT
    data_fft = data_fft * 2 / N

    # Solo se conserva la mitad positiva del espectro, por la simetría de la FFT de señales reales
    data_fft = data_fft[:N // 2, :]
    
    # Se calcula el eje de frecuencia en microHertz (µHz)
    freq = 1_000_000 * np.fft.fftfreq(N, d=granularity)[:N // 2]

    # Se construye un DataFrame con los datos FFT y se agrega la columna de frecuencia en µHz
    data_fft_df = pd.DataFrame(data_fft, columns=df.columns)
    data_fft_df.insert(0, "frequency_uHz", freq)
    
    return data_fft_df
```

```python
df_fft = data_preparation_fft(data_no_nan.drop(columns=['timestamp', 'machine_status']), 60)
```

```python
df_fft.head()
```

### Feature engineering en dominio de frecuencia

A partir de `df_fft` (espectro: magnitud vs frecuencia por sensor) se calculan 3 tipos de features para uso en clasificación o detección de fallas.

```python
# 1. ENERGÍA POR BANDAS DE FRECUENCIA
# -------------------------------------
# Se divide el eje de frecuencias en 3 bandas (bajas, medias, altas) y se calcula la energía
# (suma de magnitud²) por banda y por sensor. Las fallas suelen cambiar la distribución de
# energía entre frecuencias (ej. más energía en altas = vibración anómala). Da features
# estables y muy usadas en condición de máquinas.

sensor_cols = [c for c in df_fft.columns if c.startswith('sensor_')]
freq = df_fft['frequency_uHz'].values

# Definir bandas por tercios del rango de frecuencias (baja / media / alta)
p33, p66 = np.percentile(freq, 33), np.percentile(freq, 66)
low_mask = freq <= p33
mid_mask = (freq > p33) & (freq <= p66)
high_mask = freq > p66

energy_by_band = {}
for band_name, mask in [('low', low_mask), ('mid', mid_mask), ('high', high_mask)]:
    # Energía = suma de magnitud² en esa banda (por sensor)
    band_data = df_fft.loc[mask, sensor_cols]
    energy_by_band[band_name] = (band_data ** 2).sum(axis=0)

# Un solo vector de features: energy_{banda}_{sensor}
energy_features = {
    f'energy_{band}_{sensor}': ser[sensor]
    for band, ser in energy_by_band.items()
    for sensor in sensor_cols
}
energy_band_df = pd.DataFrame([energy_features])
print("1. Energía por bandas — shape:", energy_band_df.shape)
energy_band_df.head()
```

```python
# 2. FRECUENCIA DOMINANTE (PEAK FREQUENCY)
# ----------------------------------------
# Para cada sensor, la frecuencia (en µHz) donde la magnitud del espectro es máxima.
# En vibraciones, la frecuencia dominante suele estar ligada a giros, rodamientos, etc.;
# un cambio de esa frecuencia indica cambio de estado o falla. Muy interpretable.

dominant_freq_features = {}
for col in sensor_cols:
    idx_max = df_fft[col].idxmax()
    dominant_freq_features[f'dominant_freq_{col}'] = df_fft.loc[idx_max, 'frequency_uHz']

dominant_freq_df = pd.DataFrame([dominant_freq_features])
print("2. Frecuencia dominante por sensor — shape:", dominant_freq_df.shape)
dominant_freq_df.head()
```

```python
# 3. ESTADÍSTICAS DEL ESPECTRO POR SENSOR
# ---------------------------------------
# Por cada sensor, sobre el vector de magnitudes en frecuencia: media, desviación estándar
# y percentil 90. La media refleja el nivel global de vibración; la std y el p90 capturan
# si hay picos aislados (fallas incipientes) o si el espectro es más plano. Complementa
# energía y frecuencia dominante.

spectrum_stats_features = {}
for col in sensor_cols:
    mag = df_fft[col].values
    spectrum_stats_features[f'{col}_spectrum_mean'] = np.mean(mag)
    spectrum_stats_features[f'{col}_spectrum_std'] = np.std(mag)
    spectrum_stats_features[f'{col}_spectrum_p90'] = np.percentile(mag, 90)

spectrum_stats_df = pd.DataFrame([spectrum_stats_features])
print("3. Estadísticas del espectro (mean, std, p90) por sensor — shape:", spectrum_stats_df.shape)
spectrum_stats_df.head()
```

```python
# DataFrame único con todos los features en dominio de frecuencia (una fila = un “resumen” del espectro)
df_fft_features = pd.concat([energy_band_df.reset_index(drop=True),
                             dominant_freq_df.reset_index(drop=True),
                             spectrum_stats_df.reset_index(drop=True)], axis=1)
print("Features en dominio de frecuencia — shape:", df_fft_features.shape)
df_fft_features.head()
```

```python
def plot_fft_sensor(df, lower_freq, upper_freq, sensor_name='00-03'):
    mask = (df['frequency_uHz'] >= lower_freq) & (df['frequency_uHz'] <= upper_freq)
    plt.figure(figsize=(10, 6))
    plt.plot(df['frequency_uHz'][mask], df['sensor_00'][mask], label='sensor_00')
    plt.plot(df['frequency_uHz'][mask], df['sensor_01'][mask], label='sensor_01')
    plt.plot(df['frequency_uHz'][mask], df['sensor_02'][mask], label='sensor_02')
    plt.plot(df['frequency_uHz'][mask], df['sensor_03'][mask], label='sensor_03')
    
    plt.xlabel('Frecuencia (µHz)')
    plt.ylabel('Magnitud')
    plt.title(f'Espectro de Frecuencia para sensor {sensor_name}')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
plot_fft_sensor(df_fft, 0 ,1, '00-03')
```
